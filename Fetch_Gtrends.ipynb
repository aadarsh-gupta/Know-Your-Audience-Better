```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ["Medicare", 'Crime', 'Immigration', 'Unemployment', 'Social Security']
pytrends.build_payload(kw_list, cat=0, timeframe='now 7-d', geo = 'US-MN', gprop='')
```

```python
topics = pytrends.interest_by_region(resolution='CITY', inc_geo_code=False, inc_low_vol=True)
topics.head(20)
```

```python
topics = pytrends.interest_over_time()
topics.head(20)
```
